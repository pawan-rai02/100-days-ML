# Lets apply Gradient Descent assuming slope = 49.27 is constant and b has a starting value of -20

```python
y_pred = ((49.27 * x) + 0).reshape(6)
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color = 'red', label = 'Ols')
plt.plot(x, y_pred, color = 'green', label = 'GD')
plt.legend()
```

```python
m = 49.27
b = -20
```

**formula = -2 sigma(yi - mxi - b)**

```python
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
# learning rate
lr = .1

step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((m * x) + b).reshape(6)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color = 'red', label = 'Ols')
plt.plot(x, y_pred1, color = 'black', label = 'b = {}'.format(b))
plt.plot(x, y_pred, color = 'green', label = 'b = 0')
plt.legend()
```

```python
b = -100
m = 49.27
lr = .1

colors = ['red', 'pink', 'black', 'orange', 'green', 'blue', 'brown', 'grey', 'yellow', 'violet']

epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m * x.ravel() - b)
  step_size = loss_slope * lr
  b = b - step_size

  y_pred = ((m * x) + b).reshape(6)

  plt.plot(x, y_pred, color = colors[i], label = 'b = {}'.format(b))

plt.scatter(x, y)
```

```python
class GDRegressor:
  def __init__(self, lr = .1, epochs = 10):
    self.lr = lr
    self.epochs = epochs
    self.m = 27.82
    self.b = -120

  def fit(self, x, y):
    #calculate the b using GD
    for i in range (self.epochs):
      loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)
      self.b = self.b - (self.lr * loss_slope)
      print(loss_slope)
      print(self.b)
      print("")

    print(self.b)
```
