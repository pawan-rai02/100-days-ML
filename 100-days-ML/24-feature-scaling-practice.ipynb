```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer(as_frame=True)

df = data.frame   # includes features + target column
df.head()
```

```python
df.shape
```

```python
df['target'].sample(10)
```

```python
x = df.iloc[:, :-1]

y = df.iloc[:, -1]

y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y, test_size = .25, random_state= 41)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled  = sc.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled  = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr_scaled = LogisticRegression()
lr_scaled.fit(x_train_scaled, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred_scaled = lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

print('Before Scaling:', accuracy_score(y_test, y_pred))
print('After Scaling:', accuracy_score(y_test, y_pred_scaled))
```
