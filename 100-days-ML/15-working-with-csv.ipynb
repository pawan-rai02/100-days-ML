```python
def rename(name):
  if(name == 'Royal Challengers Bangalore'):
    return 'RCB'
  elif(name == 'Kings XI Punjab'):
    return 'KXIP'
  elif(name == 'Delhi Daredevils'):
    return 'DC'
  elif(name == 'Sunrisers Hyderabad'):
    return 'SRH'
  elif(name == 'Rising Pune Supergiants'):
    return 'RPS'
  elif(name == 'Gujarat Lions'):
    return 'GL'
  elif(name == 'Kolkata Knight Riders'):
    return 'KKR'
  elif(name == 'Mumbai Indians'):
    return 'MI'
  elif(name == 'Chennai Super Kings'):
    return 'CSK'

  else:
    return name
```

```python
ipl = pd.read_csv("/content/ipl.csv", parse_dates= ['match_date'], converters= {'team1': rename, 'team2': rename})
ipl.head()
```
