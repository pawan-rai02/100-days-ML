```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_regression(n_samples = 100, n_features = 1, n_informative = 1, n_targets= 1, noise = 20, random_state = 13)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x, y)
print(lr.coef_, lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=10)
rr.fit(x, y)
print(rr.coef_, rr.intercept_)
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(x, y)
print(rr1.coef_, rr1.intercept_)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, lr.predict(x), color = 'red', label = 'Linear Regression')
plt.plot(x, rr.predict(x), color = 'green', label = 'Ridge Regression')
plt.plot(x, rr1.predict(x), color = 'yellow', label = 'Ridge Regression 100')
plt.legend()
```
