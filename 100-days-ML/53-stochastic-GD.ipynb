```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
# Load the diabetes dataset

x, y = load_diabetes(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state = 2)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test, y_pred)
```

```python
class SGDRegressor:
  def __init__(self, lr = .01, epochs = 100):
    self.lr = lr
    self.epochs = epochs
    self.coeff_ = None
    self.intercept_ = None

  def fit(self, x_train, y_train):
    #init coeff
    self.intercept_ = 0
    self.coeff_ = np.ones(x_train.shape[1])

    for i in range(self.epochs):
      for j in range(x_train.shape[0]):
        idx = np.random.randint(0, x_train.shape[0])
        y_hat = np.dot(x_train[idx], self.coeff_) + self.intercept_

        intercept_der = -2 * (y_train[idx] - y_hat)
        self.intercept_ = self.intercept_ - (self.lr * intercept_der)

        coeff_der = -2 * np.dot((y_train[idx] - y_hat), x_train[idx])
        self.coeff_ = self.coeff_ - (self.lr * intercept_der * x_train[idx])

    print(self.intercept_, self.coeff_)

  def predict(self, x_test):
    return np.dot(x_test, self.coeff_) + self.intercept_
```

```python
sgd = SGDRegressor(epochs = 50, lr = .01)
```

```python
sgd.fit(x_train, y_train)
```

```python
y_pred = sgd.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
