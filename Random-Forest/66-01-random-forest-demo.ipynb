```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
```

```python
x, y = make_classification(n_features=5, n_informative= 5, n_redundant= 0, n_clusters_per_class=1, random_state=42)
```

```python
df = pd.DataFrame(x, columns=[f'feature_{i}' for i in range(x.shape[1])])
df['target'] = y
print(df.shape)
sns.pairplot(df, hue='target')
```

```python
import random
```

```python
# Function to sample random rows from a DataFrame
def sample_rows(df, percent):
    return df.sample(int(percent * df.shape[0]), replace=True)
```

```python
# Function to sample random features from a DataFrame
def sample_features(df, percent):
    cols = random.sample(df.columns.tolist()[:-1], int(percent * len(df.columns)))
    return df[cols]
```

```python
# function for combined sampling of rows and features
def sample_rows_and_features(df, row_percent, feature_percent):
    sampled_rows = sample_rows(df, row_percent)
    sampled_features = sample_features(sampled_rows, feature_percent)
    return sampled_features
```

```python
df1 = sample_rows(df, 0.1)
```

```python
df2 = sample_rows(df, 0.1)
df3 = sample_rows(df, 0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
clf1.fit(df1.drop('target', axis=1), df1['target'])
clf2.fit(df2.drop('target', axis=1), df2['target'])
clf3.fit(df3.drop('target', axis=1), df3['target'])
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
