```python
from sklearn.datasets import make_classification
import numpy as np
x, y = make_classification(n_samples = 100, n_features= 2, n_informative= 1, n_redundant= 0, n_classes= 2, n_clusters_per_class= 1, random_state= 42, hypercube= False, class_sep = 10)
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap= 'winter', s=100)
```

```python
def step(x):
  return 1 if x > 0 else 0
```

```python
def perceptron(x, y):

  x = np.insert(x, 0, 1, 1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0, 100)
    y_hat = step(np.dot(weights, x[j]))
    if y_hat != y[j]:
      weights += lr *(y[j] - y_hat) * x[j]

  return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(x, y)
```

```python
print(intercept_, coef_)
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, 'r', linewidth= 3)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap= 'winter', s= 100)
plt.ylim(-5, 8) # Adjusted y-axis limits
plt.xlabel('Feature 1') # Added x-axis label
plt.ylabel('Feature 2') # Added y-axis label
plt.title('Perceptron Decision Boundary') # Added a title
plt.grid(True) # Added a grid
plt.show() # Explicitly show the plot
```
