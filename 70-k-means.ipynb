```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_blobs

# Generate data with 3 clusters
X, y = make_blobs(n_samples=400, centers=4, n_features=2, random_state=42, cluster_std=1.0)

# Update the dataframe with the new clustered data
df = pd.DataFrame(X, columns=['cgpa', 'iq'])
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'], df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters= i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
```

```python
x = df.iloc[:, :].values
km = KMeans(n_clusters= 4)
y_means = km.fit_predict(x)
y_means
```

```python
x[y_means == 0, 0]
```

```python
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s = 100, c = 'purple', label = 'Cluster 4')
```
